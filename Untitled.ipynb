# Sales Analysis

```python
import pandas as pd
import os
```

#### Task-1: Merge the 12 months of sales data into single csv file

```python
files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index=False) #merge all different months file into one
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head() 
```

### Clean up the data!

```python
 #drop rows of NAN
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how='all')
all_data.head()
```

#### Find 'Or' and delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2] !='Or']
```

##### convert column to correct type


```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # make float
```

  ### Augment data with additional columns

#### Task 2: Add Month Column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

#### Task:3 Add a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

```python
all_data
```

#### Add a city column

```python
# let's use .apply()

all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1] + x.split(',')[2].split(' ')[1])

all_data.head()
```

#### Question 4: What products are most often sold together?

```python
# finding duplicate values in our dataframe
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(100)
```

```python
# Count most common pairs of products bought together
from itertools import combinations
from collections import Counter

count = Counter()  # store frequency of product pairs

for row in df['Grouped']:
    row_list = row.split(',')  # split products in each order
    count.update(Counter(combinations(row_list, 2)))  # count all 2-item combos

# display top 10 most frequent product pairs
for key, value in count.most_common(10):
    print(key, value)
```

```python
product_group = all_data.groupby('Product').sum(numeric_only=True)

quantity_ordered = product_group['Quantity Ordered']

products = list(product_group.index)  # extract product names as list

plt.bar(products, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(products, rotation = 'vertical', size=8);
```

```python
# Group data: mean price per product
prices = all_data.groupby('Product').mean(numeric_only=True)['Price Each']
# Total quantity ordered per product
quantity_ordered = all_data.groupby('Product').sum(numeric_only=True)['Quantity Ordered']
# List of product names
products = list(quantity_ordered.index)

# Create plot
fig, ax1 = plt.subplots()

# Bar plot for Quantity Ordered
ax1.bar(products, quantity_ordered, color='g')
ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')

# Second y-axis for Price
ax2 = ax1.twinx() # twinx() is create second axis sharing same x-axis
ax2.plot(products, prices, 'b-')
ax2.set_ylabel('Price ($)', color='b')

# Fix tick labels (set ticks first)
ax1.set_xticks(range(len(products)))
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
